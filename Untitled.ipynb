```python
import threading
import queue
import re
import time
import sounddevice as sd
import numpy as np
import torch
from transformers import AutoTokenizer
from parler_tts import ParlerTTSForConditionalGeneration
from openai import OpenAI

client = OpenAI(
    base_url="https://api.sambanova.ai/v1", 
    api_key="4a81daa9-5f3d-409b-9f30-ebedb379219a"
)

# Stream LLM output and split into sentences, alternating between two queues.
def stream_llm_to_sentence_queues(messages, q1, q2):
    response = client.chat.completions.create(
        model="Meta-Llama-3.2-3B-Instruct",
        messages=messages,
        stream=True
    )
    buffer = ""
    toggle = 0
    sentence_end_pattern = re.compile(r'([।.?!])')

    for chunk in response:
        delta = chunk.choices[0].delta
        if hasattr(delta, "content") and delta.content:
            text = delta.content
            buffer += text
            while True:
                m = sentence_end_pattern.search(buffer)
                if not m:
                    break
                end_idx = m.end(1)
                sentence = buffer[:end_idx].strip()
                buffer = buffer[end_idx:]
                if sentence:
                    if toggle == 0:
                        q1.put(sentence)
                    else:
                        q2.put(sentence)
                    toggle = 1 - toggle

    if buffer.strip():
        sentence = buffer.strip()
        if toggle == 0:
            q1.put(sentence)
        else:
            q2.put(sentence)
    q1.put(None)
    q2.put(None)

# TTS worker

def tts_worker(text_queue, audio_queue, model, tokenizer, desc_input_ids, desc_attention_mask, sampling_rate):
    while True:
        sentence = text_queue.get()
        if sentence is None:
            break
        print(f"[TTS Worker] Received sentence: {sentence}")
        prompt_inputs = tokenizer(sentence, return_tensors="pt")
        prompt_input_ids = prompt_inputs.input_ids.to(model.device)
        prompt_attention_mask = prompt_inputs.attention_mask.to(model.device)
        with torch.no_grad():
            speech_output = model.generate(
                input_ids=desc_input_ids,
                attention_mask=desc_attention_mask,
                prompt_input_ids=prompt_input_ids,
                prompt_attention_mask=prompt_attention_mask
            )
        if speech_output is None or speech_output.nelement() == 0:
            print(f"[TTS Worker] Empty audio generated for: {sentence}")
            continue
        audio = speech_output.cpu().numpy().squeeze()
        if audio.ndim > 1:
            audio = audio.flatten()
        audio = audio.astype(np.float32)
        max_val = np.max(np.abs(audio))
        print(f"[TTS Worker] Audio length={len(audio)} max amplitude={max_val:.5f}")
        if max_val > 0:
            audio = audio / max_val
        chunk_size = int(0.5 * sampling_rate)
        start = 0
        total_len = len(audio)
        while start < total_len:
            end = min(start + chunk_size, total_len)
            chunk_audio = audio[start:end]
            print(f"[TTS Worker] Sending chunk {start}:{end} to audio queue")
            audio_queue.put(chunk_audio)
            start = end
            chunk_size *= 2

# Playback thread

def playback_worker(p1, p2, done_event, sampling_rate):
    next_from_p1 = True
    while True:
        played = False
        if next_from_p1:
            if not p1.empty():
                chunk = p1.get()
                print(f"[Playback] Playing chunk from P1, len={len(chunk)}")
                sd.play(chunk, samplerate=sampling_rate)
                sd.wait()
                next_from_p1 = False
                played = True
            elif not p2.empty():
                chunk = p2.get()
                print(f"[Playback] Playing chunk from P2, len={len(chunk)}")
                sd.play(chunk, samplerate=sampling_rate)
                sd.wait()
                played = True
        else:
            if not p2.empty():
                chunk = p2.get()
                print(f"[Playback] Playing chunk from P2, len={len(chunk)}")
                sd.play(chunk, samplerate=sampling_rate)
                sd.wait()
                next_from_p1 = True
                played = True
            elif not p1.empty():
                chunk = p1.get()
                print(f"[Playback] Playing chunk from P1, len={len(chunk)}")
                sd.play(chunk, samplerate=sampling_rate)
                sd.wait()
                played = True
        if played:
            continue
        if done_event.is_set() and p1.empty() and p2.empty():
            print("[Playback] All queues empty and done event set. Exiting.")
            break
        time.sleep(0.01)

# Main pipeline

def main():
    Q1 = queue.Queue()
    Q2 = queue.Queue()
    P1 = queue.Queue()
    P2 = queue.Queue()

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model1 = ParlerTTSForConditionalGeneration.from_pretrained("ai4bharat/indic-parler-tts").to(device)
    model2 = ParlerTTSForConditionalGeneration.from_pretrained("ai4bharat/indic-parler-tts").to(device)
    tokenizer = AutoTokenizer.from_pretrained("ai4bharat/indic-parler-tts")
    description_tokenizer = AutoTokenizer.from_pretrained(model1.config.text_encoder._name_or_path)

    description = ("Rohit's voice is monotone yet slightly fast in delivery, with a very close recording that almost has no background noise.")
    desc_inputs = description_tokenizer(description, return_tensors="pt").to(device)
    desc_input_ids = desc_inputs.input_ids
    desc_attention_mask = desc_inputs.attention_mask

    tts_thread1 = threading.Thread(
        target=tts_worker,
        args=(Q1, P1, model1, tokenizer, desc_input_ids, desc_attention_mask, model1.config.sampling_rate)
    )
    tts_thread2 = threading.Thread(
        target=tts_worker,
        args=(Q2, P2, model2, tokenizer, desc_input_ids, desc_attention_mask, model2.config.sampling_rate)
    )
    tts_thread1.start()
    tts_thread2.start()

    done_event = threading.Event()
    playback_thread = threading.Thread(
        target=playback_worker,
        args=(P1, P2, done_event, model1.config.sampling_rate)
    )
    playback_thread.start()

    system_message = {"role": "system", "content": "You are AI voice assistant, Keep your answer very short, without abbreviation. Ensure that you Only answer in hindi text only"}
    user_query = input("Enter your question in Hindi: ")
    user_message = {"role": "user", "content": user_query}
    messages = [system_message, user_message]

    stream_llm_to_sentence_queues(messages, Q1, Q2)
    print("[Main] Finished streaming sentences. Waiting for audio queues to drain...")

    tts_thread1.join()
    tts_thread2.join()

    while not P1.empty() or not P2.empty():
        time.sleep(0.1)

    done_event.set()
    playback_thread.join()

    print("Streaming speech synthesis complete.")

if __name__ == "__main__":
    main()
```

```python
import threading
import asyncio
import queue
import time
import sounddevice as sd
import numpy as np
import torch
import warnings
import os
import sys
import contextlib

# Suppress warnings and ALSA errors
os.environ["PYTHONWARNINGS"] = "ignore"
warnings.filterwarnings("ignore")


from transformers import AutoTokenizer
from parler_tts import ParlerTTSForConditionalGeneration
from openai import OpenAI
import re

# === Initialization Block (run once) ===
client = OpenAI(
    base_url="https://api.sambanova.ai/v1", 
    api_key="4a81daa9-5f3d-409b-9f30-ebedb379219a"
)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
TOKENIZER = AutoTokenizer.from_pretrained("ai4bharat/indic-parler-tts")
MODEL_1 = ParlerTTSForConditionalGeneration.from_pretrained("ai4bharat/indic-parler-tts").to(DEVICE)
MODEL_2 = ParlerTTSForConditionalGeneration.from_pretrained("ai4bharat/indic-parler-tts").to(DEVICE)
DESC_TOKENIZER = AutoTokenizer.from_pretrained(MODEL_1.config.text_encoder._name_or_path)
DESCRIPTION = ("Rohit's voice is monotone yet slightly fast in delivery, with a very close recording that almost has no background noise.")
DESC_INPUTS = DESC_TOKENIZER(DESCRIPTION, return_tensors="pt").to(DEVICE)
DESC_INPUT_IDS = DESC_INPUTS.input_ids
DESC_ATTN_MASK = DESC_INPUTS.attention_mask
SAMPLING_RATE = MODEL_1.config.sampling_rate
sd.default.latency = 'high'

MAX_CHUNK_CHARS = 10000

# === Async Playback Queue ===
async def async_playback(queue1, queue2, done_event, initial_chunk_chars, latency_event):
    next_from_queue1 = True
    chunk_chars = initial_chunk_chars
    latency_measured = False

    while True:
        try:
            if next_from_queue1:
                audio = await queue1.get()
                next_from_queue1 = False
            else:
                audio = await queue2.get()
                next_from_queue1 = True
        except asyncio.CancelledError:
            break

        if latency_event and not latency_measured:
            latency_event.set()
            latency_measured = True

        with suppress_stderr():
            sd.play(audio, samplerate=SAMPLING_RATE)
            sd.wait()

        chunk_chars = min(chunk_chars * 2, MAX_CHUNK_CHARS)

        if done_event.is_set() and queue1.empty() and queue2.empty():
            break

# === Core Worker Functions ===
def tts_worker(text_queue, audio_queue, model, loop):
    async def put_audio(audio):
        await audio_queue.put(audio)

    while True:
        text = text_queue.get()
        if text is None:
            break
        print(f"🤖 LLM: {text}")
        prompt_inputs = TOKENIZER(text, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            speech_output = model.generate(
                input_ids=DESC_INPUT_IDS,
                attention_mask=DESC_ATTN_MASK,
                prompt_input_ids=prompt_inputs.input_ids,
                prompt_attention_mask=prompt_inputs.attention_mask
            )
        audio = speech_output.cpu().numpy().squeeze()
        if audio.ndim > 1:
            audio = audio.flatten()
        audio = audio.astype(np.float32)
        max_val = np.max(np.abs(audio))
        if max_val > 0:
            audio = audio / max_val

        asyncio.run_coroutine_threadsafe(put_audio(audio), loop)


def stream_llm_to_chunks(prompt, q1, q2, initial_chunk_chars):
    buffer = ""
    toggle = 0
    chunk_chars = initial_chunk_chars
    response = client.chat.completions.create(
        model="Meta-Llama-3.2-3B-Instruct",
        messages=[
            {"role": "system", "content": "You are a Hindi-only assistant. Keep responses short."},
            {"role": "user", "content": prompt}
        ],
        stream=True
    )
    for chunk in response:
        delta = chunk.choices[0].delta
        if hasattr(delta, "content") and delta.content:
            buffer += delta.content
            if len(buffer) >= chunk_chars:
                if toggle == 0:
                    q1.put(buffer.strip())
                else:
                    q2.put(buffer.strip())
                buffer = ""
                toggle = 1 - toggle
                chunk_chars = min(chunk_chars * 2, MAX_CHUNK_CHARS)
    if buffer:
        if toggle == 0:
            q1.put(buffer.strip())
        else:
            q2.put(buffer.strip())
    q1.put(None)
    q2.put(None)


def speak_from_prompt(prompt: str, initial_chunk_chars: int = 20, measure_latency: bool = False):
    Q1, Q2 = queue.Queue(), queue.Queue()
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)
    A1, A2 = asyncio.Queue(), asyncio.Queue()
    done_event = threading.Event()
    latency_event = threading.Event() if measure_latency else None

    loop_thread = threading.Thread(target=loop.run_forever)
    loop_thread.start()

    tts_thread1 = threading.Thread(target=tts_worker, args=(Q1, A1, MODEL_1, loop))
    tts_thread2 = threading.Thread(target=tts_worker, args=(Q2, A2, MODEL_2, loop))
    tts_thread1.start()
    tts_thread2.start()

    start_time = time.time() if measure_latency else None
    asyncio.run_coroutine_threadsafe(async_playback(A1, A2, done_event, initial_chunk_chars, latency_event), loop)

    stream_llm_to_chunks(prompt, Q1, Q2, initial_chunk_chars)
    tts_thread1.join()
    tts_thread2.join()
    done_event.set()

    while not A1.empty() or not A2.empty():
        time.sleep(0.1)

    loop.call_soon_threadsafe(loop.stop)
    loop_thread.join()

    if measure_latency and latency_event:
        latency_event.wait()
        latency = time.time() - start_time
        print(f"✅ Speech synthesis complete.\n⏱️ First audio latency: {latency:.2f} seconds")
    else:
        print("✅ Speech synthesis complete.")
```

```python
#This lags between each output


import threading
import asyncio
import queue
import time
import sounddevice as sd
import numpy as np
import torch
import warnings
import os
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

from transformers import AutoTokenizer
from parler_tts import ParlerTTSForConditionalGeneration
from openai import OpenAI

os.environ["PYTHONWARNINGS"] = "ignore"
warnings.filterwarnings("ignore")

client = OpenAI(
    base_url="https://api.sambanova.ai/v1",
    api_key="4a81daa9-5f3d-409b-9f30-ebedb379219a"
)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
TOKENIZER = AutoTokenizer.from_pretrained("ai4bharat/indic-parler-tts")
MODEL_1 = ParlerTTSForConditionalGeneration.from_pretrained("ai4bharat/indic-parler-tts").to(DEVICE)
MODEL_2 = ParlerTTSForConditionalGeneration.from_pretrained("ai4bharat/indic-parler-tts").to(DEVICE)

DESC_TOKENIZER = AutoTokenizer.from_pretrained(MODEL_1.config.text_encoder._name_or_path)
DESCRIPTION = "Rohit's voice is monotone yet slightly fast in delivery."
DESC_INPUTS = DESC_TOKENIZER(DESCRIPTION, return_tensors="pt").to(DEVICE)
DESC_INPUT_IDS = DESC_INPUTS.input_ids
DESC_ATTN_MASK = DESC_INPUTS.attention_mask

SAMPLING_RATE = MODEL_1.config.sampling_rate
sd.default.latency = 'low'

MAX_CHUNK_CHARS = 10000

def tts_worker(text_queue, audio_queue, model, loop):
    async def put_audio(audio):
        await audio_queue.put(audio)

    while True:
        text = text_queue.get()
        if text is None:
            break

        prompt_inputs = TOKENIZER(text, return_tensors="pt").to(DEVICE)
        try:
            # Sanity check
            if prompt_inputs.input_ids.max().item() >= model.config.vocab_size:
                print("⚠️ Token out of vocab range.")
                continue

            with torch.no_grad():
                output = model.generate(
                    input_ids=DESC_INPUT_IDS,
                    attention_mask=DESC_ATTN_MASK,
                    prompt_input_ids=prompt_inputs.input_ids,
                    prompt_attention_mask=prompt_inputs.attention_mask,
                    return_dict_in_generate=False
                )

            audio = output.cpu().numpy().squeeze()
            if audio.ndim > 1:
                audio = audio.flatten()
            audio = audio.astype(np.float32)
            max_val = np.max(np.abs(audio))
            if max_val > 0:
                audio = audio / max_val

            asyncio.run_coroutine_threadsafe(put_audio(audio), loop)

        except RuntimeError as e:
            print(f"❌ CUDA RuntimeError: {e}")
        except Exception as e:
            print(f"❌ TTS generation error: {e}")


async def async_playback(queue1, queue2, done_event, latency_event):
    latency_measured = False
    toggle = True

    while True:
        try:
            audio = None
            queue_name = None

            if toggle:
                audio = await queue1.get()
                queue_name = "[Q1]"
                toggle = False
            else:
                audio = await queue2.get()
                queue_name = "[Q2]"
                toggle = True

            if audio is None:
                if done_event.is_set() and queue1.empty() and queue2.empty():
                    break
                continue

            if latency_event and not latency_measured:
                latency_measured = True
                latency_event.set()

            print(f"{queue_name} ▶️ Playing {len(audio)} samples")
            sd.play(audio, samplerate=SAMPLING_RATE)
            sd.wait()

        except asyncio.CancelledError:
            break

    print("🔁 Playback finished.")

def stream_llm_to_chunks(prompt, q1, q2, small_chunk=20, large_chunk=60, max_queue_size=5, first_chunk_ready=None):
    buffer = ""
    toggle = 0
    chunk_sizes = [small_chunk, large_chunk]
    chunk_index = 0

    response = client.chat.completions.create(
        model="Meta-Llama-3.2-3B-Instruct",
        messages=[
            {"role": "system", "content": "You are a Hindi-only assistant."},
            {"role": "user", "content": prompt}
        ],
        stream=True
    )

    for chunk in response:
        delta = chunk.choices[0].delta
        if hasattr(delta, "content") and delta.content:
            buffer += delta.content
            print(delta.content, end="", flush=True)
            current_chunk_size = chunk_sizes[chunk_index % 2]

            while len(buffer) >= current_chunk_size:
                chunk_text = buffer[:current_chunk_size].strip()
                target_queue = q1 if toggle == 0 else q2

                while target_queue.qsize() >= max_queue_size:
                    time.sleep(0.01)
                target_queue.put(chunk_text)

                if first_chunk_ready and not first_chunk_ready.is_set():
                    first_chunk_ready.set()

                buffer = buffer[current_chunk_size:]
                toggle = 1 - toggle
                chunk_sizes[chunk_index % 2] = min(chunk_sizes[chunk_index % 2] * 2, MAX_CHUNK_CHARS)
                chunk_index += 1

    if buffer.strip():
        target_queue = q1 if toggle == 0 else q2
        while target_queue.qsize() >= max_queue_size:
            time.sleep(0.01)
        target_queue.put(buffer.strip())

        if first_chunk_ready and not first_chunk_ready.is_set():
            first_chunk_ready.set()

    q1.put(None)
    q2.put(None)

def speak_from_prompt(prompt: str, small_chunk=20, large_chunk=60, measure_latency=False):
    Q1, Q2 = queue.Queue(), queue.Queue()
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)

    # ✅ Use bounded asyncio queues for audio buffering
    A1, A2 = asyncio.Queue(maxsize=5), asyncio.Queue(maxsize=5)
    
    done_event = threading.Event()
    latency_event = threading.Event() if measure_latency else None
    first_chunk_ready = threading.Event()

    loop_thread = threading.Thread(target=loop.run_forever)
    loop_thread.start()

    # Start TTS workers
    tts_thread1 = threading.Thread(target=tts_worker, args=(Q1, A1, MODEL_1, loop))
    tts_thread2 = threading.Thread(target=tts_worker, args=(Q2, A2, MODEL_2, loop))
    tts_thread1.start()
    tts_thread2.start()

    # Start LLM stream
    stream_thread = threading.Thread(
        target=stream_llm_to_chunks,
        args=(prompt, Q1, Q2, small_chunk, large_chunk, 5, first_chunk_ready)
    )
    stream_thread.start()

    # ⏳ Wait for first chunk before playing
    first_chunk_ready.wait()

    # ✅ Start playback: will only consume already prepared audio
    playback_future = asyncio.run_coroutine_threadsafe(
        async_playback(A1, A2, done_event, latency_event), loop
    )

    stream_thread.join()
    tts_thread1.join()
    tts_thread2.join()

    done_event.set()
    playback_future.result()

    loop.call_soon_threadsafe(loop.stop)
    loop_thread.join()

    if measure_latency and latency_event:
        latency_event.wait()
        latency = time.time() - start_time
        print(f"\n✅ Speech synthesis complete.\n⏱️ First audio latency: {latency:.2f} seconds")
    else:
        print("\n✅ Speech synthesis complete.")
```
